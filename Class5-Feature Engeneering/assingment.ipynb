```python
li = [110,105,115,120,110,130,150,100,105]
import pandas as pd
df = pd.DataFrame(li, columns=["Price"])
df
```

```python
min_price = min(df['Price'])
min_price
```

```python
max_price = max(df['Price'])
max_price
```

```python
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler()
df['Normalization'] = norm.fit_transform(df[['Price']])
df
```

```python
from sklearn.preprocessing import StandardScaler
stand = StandardScaler()
df['StandardScaler'] = stand.fit_transform(df[['Price']])
df
```

```python
from sklearn.preprocessing import FunctionTransformer
logTrans = FunctionTransformer()
df['Log Transform'] = logTrans.fit_transform(df[['Price']])
df
```

```python
from sklearn.preprocessing import RobustScaler
RoSc=RobustScaler()
df['Robust Scaler'] = RoSc.fit_transform(df[['Price']])
df
```

```python
from sklearn.preprocessing import MaxAbsScaler
Maxscaler = MaxAbsScaler()
df['Max Absolute Scaler'] = Maxscaler.fit_transform(df[['Price']])
df
```

```python
df.to_excel('Feature.xlsx')
```
